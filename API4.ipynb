```python
import requests
from pprint import PrettyPrinter
pp = PrettyPrinter()
```

```python
MY_API_KEY = '2592e729'
```

```python
url = 'http://www.omdbapi.com/?apikey=' + MY_API_KEY
year = ''
movie_name = "Avengers"
params = {
    't':movie_name,
    'type':'movie',
    'y':year,
    'plot':'full'
}

```

```python
response = requests.get(url,params=params).json()
pp.pprint(response)
```

```python
response['Poster']
```

```python
url = response['Poster']
r = requests.get(url, allow_redirects=True)
filename = response['Title']+'.jpg'
open(filename, 'wb').write(r.content)
```
